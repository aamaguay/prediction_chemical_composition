```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error
```

```python
df = pd.read_csv('/content/drive/My Drive/Colab Notebooks/data_final_Na.csv',sep=';').iloc[:,2:]
df_train = df[df.id_train == 1].loc[ : , df.columns != 'id_train']
df_test = df[df.id_train == 0].loc[ : , df.columns != 'id_train']
```

```python
import h2o
h2o.init(nthreads= -1)
```

```python
h2o_train = h2o.H2OFrame(df_train)
h2o_test = h2o.H2OFrame(df_test)
```

```python
h2o_train.head()
```

# Form list features

```python
features = list(h2o_train.columns)
features.remove('Na')
```

# Split train & validation data set

```python
df_split = h2o_train.split_frame(ratios= [0.85], seed = 12)
new_train = df_split[0]
new_valid = df_split[1]
```

# Import xgb algorithm


```python
from h2o.estimators import H2OXGBoostEstimator
```

# Test with default parameters

```python
xgb_default = H2OXGBoostEstimator()
xgb_default.train(x = features ,
                  y= 'Na',
                  training_frame = new_train )
```

# Evaluate perfomance in Validation dataset, then predict values 

```python
xgb_default.model_performance(new_valid)
```

```python
yhat_xgb_default = xgb_default.predict(new_valid)
```

# XGB with cross validation with early stopping and Full grid search

```python
from h2o.grid.grid_search import H2OGridSearch
```

```python
search_criteria = {'strategy': 'Cartesian'}
```

```python
hyper_params = { 'learn_rate': [0.01,0.1,0.5,0.9],
                'reg_alpha' : [0.1, 1, 10],
                'subsample' : [0.4, 0.6],
                'col_sample_rate_per_tree' : [0.4, 0.7 ],
                'max_depth' : [5, 10, 12]
                #'sample_rate':[0.7, 0.8],
                #'col_sample_rate': [0.4, 0.7],
                #'ntrees': [ 90, 100], 
                #'max_depth': [ 30]
}
```

```python
xgb_full_grid = H2OGridSearch(
    H2OXGBoostEstimator(model_id= 'xgb_1st_manual_cv_es_fullgrid', seed= 123,
                                 #colsample_bynode=.5,
                                 #col_sample_rate_per_tree=.7,
                                  #score_tree_interval = 1,
                                   #learn_rate=0.01,
                                 distribution= "auto",
                                 min_split_improvement=1e-3,
                                 nfolds = 5,
                                 stopping_metric= 'rmse',
                                 stopping_rounds = 12,
                                 ),
                              search_criteria = search_criteria,
                              hyper_params = hyper_params
    
)
```

```python
xgb_full_grid.train(x = features ,
                  y= 'Na',
                  training_frame = h2o_train )
```

```python
xgb_full_grid_sorted = xgb_full_grid.get_grid(sort_by= 'rmse', decreasing = False) 
```

```python
best_model_id = xgb_full_grid_sorted.model_ids[0]
best_xgb_tunning_model = h2o.get_model(best_model_id )
best_xgb_tunning_model.summary()
```

```python
best_xgb_tunning_model.model_performance(h2o_test)
```

```python
xgb_full_grid.get_grid(sort_by= 'rmse', decreasing = False) 
```

```python
model_path_full_grid = h2o.save_model(model= best_xgb_tunning_model, force=True)
# Grid_XGBoost_Key_Frame__upload_8495eec87d34cb459762c0833f1a495d.hex_model_python_1638232654351_1_model_124
```

```python
pd.DataFrame(best_xgb_tunning_model.actual_params.items(), columns =['parameter', 'value']).set_index('parameter').transpose()
```
